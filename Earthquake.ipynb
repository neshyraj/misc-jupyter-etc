```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pycountry as pyc
import warnings
warnings.filterwarnings("ignore") 
```

```python
#read in data and convert place variable to only contain string values
data = pd.read_excel("Mag6PlusEarthquakes_1900-2013.xlsx")

for i in np.arange(0, len(data['place']), 1):
    if isinstance(data['place'][i], float):
        data['place'][i] = str(data['place'][i])
    else:
        next

data.head(2)
```

```python
#create new column that contains country information
countries = data['place'].copy()

for i in np.arange(0, len(data['place']), 1):
    for country in pyc.countries:
        if country.name in data['place'][i]:
            countries[i] = country.name
```

```python
#manually adjust countries since russia, taiwan, iran are not in pythons dictionary
data['country'] = countries

for i in np.arange(0, len(data['place']), 1):
    if 'Russia' in data['place'][i]:
        data['country'][i] = 'Russia'
    elif 'Taiwan' in data['place'][i]:
        data['country'][i] = 'Taiwan'
    elif 'Iran' in data['place'][i]:
        data['country'][i] = 'Iran'
    else:
        next

data.head(3)
```

```python
#get land or water
from mpl_toolkits.basemap import Basemap
bm = Basemap()

terrain = []
for i in np.arange(0, len(data['latitude']), 1):
    if bm.is_land(data['longitude'][i], data['latitude'][i]):
        terrain.append('Land')
    else:
        terrain.append('Water')
```

```python
data['terrain'] = terrain
data.head(3)
```
